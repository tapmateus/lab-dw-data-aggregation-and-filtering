This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by first performing data cleaning, formatting, and structuring.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
df.columns
```

1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

```python
#filtering for both total > 1000 & response == 'Yes'
newdf = df[(df['Total Claim Amount'] > 1000) & (df['Response'] == 'Yes')]
newdf
```

2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.

```python
## new df filtered the original to retrieve only customer who responded 'yes'
newdf2 = df[(df['Response'] == 'Yes')]

##average total claim amout for each policy type using group by
newdf2.groupby('Policy').agg({'Total Claim Amount':'mean'})

#overal personal policy insurance have a total claim higher than corporate (i would expect the contrary) and special
#even higher than both
```

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# grouping and counting each policy in each state

df.groupby('State').agg({'Policy':'count'}) #every state has more than 500 customers so the're already included
```

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
#finding only for gender
df.groupby('Gender').agg({'Customer Lifetime Value':['min','max','mean']})

#only for education
df.groupby('Education').agg({'Customer Lifetime Value':['min','max','mean']})

#gender in each education level
df.groupby(['Education','Gender']).agg({'Customer Lifetime Value':['min','max','median']})
```
